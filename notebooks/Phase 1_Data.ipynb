```python
def f1_from_text(sample):
  title = sample['title'].lower() # lowercases every title for every sample
  return any(keyword in title for keyword in f1_keywords)#runs through every keyword in f1_keywords and checks their presence in title

filtered_dataset = dataset.filter(f1_from_text)
```

```python
for i in range(0, len(filtered_dataset), len(filtered_dataset)//20):
    print(filtered_dataset[i]['title'])
```

```python
import re

def clean_text(text):
    # 1. Cut everything from reference/external sections onwards
    text = re.sub(r'==\s*(References|External links|See also|Further reading|Notes|Bibliography)\s*==.*', '', text, flags=re.DOTALL|re.IGNORECASE)

    # 2. Remove citation markers like [1], [23], [citation needed]
    text = re.sub(r'\[[^\]]*\]', '', text)

    # 3. Normalize whitespace
    text = re.sub(r'\n{3,}', '\n\n', text)
    text = re.sub(r' {2,}', ' ', text)

    return text.strip()
```

```python
corpus = "<|endoftext|>".join(clean_text(item['text']) for item in filtered_dataset)
```

```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokens = tokenizer.encode(corpus)
print(f"Exact token count: {len(tokens):,}")
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/GPT2 Task on F1 Data/data/f1_corpus.txt', 'w', encoding='utf-8') as f:
    f.write(corpus)

print("Saved successfully")
print(f"File size: {len(corpus) / 1024 / 1024:.2f} MB")
```
